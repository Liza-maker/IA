```python
def link_to_mp4(link):
    """
    Fonction qui télécharge une vidéo youtube en .mp4 (vérifie automatiquement si vidéo du même titre déjà téléchargée)

    Args:
        link (str): lien de la vidéo youtube
    
    Sortie : Téléchargement de la vidéo
    """
    #Vérifie si le lien est bien youtube (source regex : https://stackoverflow.com/questions/71000139/javascript-regex-for-youtube-video-and-shorts-id)
    pattern = r'(youtu.*be.*)\/(watch\?v=|embed\/|v|shorts|)(.*?((?=[&#?])|$))'
    link_verify = re.findall(pattern,link)
    if not link_verify:
        return "Erreur dans le format de la vidéo"
    # Vérifie si link est une liste et le formatte : 
    if isinstance(link,list):
        link = [link]
    with YoutubeDL() as ydl:
        ydl.download(link)

# Test :
link = "https://www.youtube.com/watch?v=C6EgYpH9nmM"
#link_to_mp4(link)
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from sentence_transformers import SentenceTransformer
from pypdf import PdfReader

def readPdf(list_file):
    """
    Fonction qui transforme des pdf en texte :

    Args:
        file (list): liste des fichiers

    Returns:
        text : contenu du fichier
    """
    list_texts = []
    for file in list_file:
        reader = PdfReader(file)
        nb_pages = len(reader.pages)
        for i in range(nb_pages):
            page = reader.pages[i]
            text = page.extract_text()
            list_texts.append(text)
    return list_texts

def chunks(s, k):
    """
    Fonction qui prend du texte et le découpe en chunk de k taille

    Args:
        s (text): texte à découper
        k (int): découpage en k chunks

    Returns:
        list_texts : liste de liste texte découpé
    """
    return [s[i:i+k] for i in range(0, len(s), k)]


document = readPdf(["pdf-exemple.pdf","sample.pdf","rugby.pdf"])
TAILLE_CHUNK = 250
#On découpe chaque document en des chunks de 50 (fait liste de liste)
stck_doc = []
for file in document:
    stck_doc.append(chunks(file,TAILLE_CHUNK))

#On retire la double liste de stck_doc :
unique_list = []
for i in range(len(stck_doc)):
    for j in range(len(stck_doc[i])):
        unique_list.append(stck_doc[i][j])
print(unique_list)


#On embedding (transférer les phrases par des vecteurs)
modelEmbedding = SentenceTransformer("all-MiniLM-L6-v2")
embedding = modelEmbedding.encode(unique_list, show_progress_bar=True)
print(embedding)
print("Longueur embedding 1 : ",len(embedding[0]))
```

```python
from sentence_transformers import SentenceTransformer, util

user_text = input("Entrez votre phrase : ")
modelEmbedding = SentenceTransformer("all-MiniLM-L6-v2")
embedding_user = modelEmbedding.encode(user_text,show_progress_bar=True)
print(embedding_user)
print("Longueur du vecteur : ",len(embedding_user))
```

```python
import math

#On utilise la recherche sémantique pour retrouver les vecteurs les plus proches (donc les plus probables)
result = util.semantic_search(embedding_user,embedding)[0]

output_text = []

#On inclut une réponse par défaut si la similarité est pas assez haute pour répondre à la question posée.

for item in result:
    score = math.ceil(item["score"]*100)/100
    print(score, "|", unique_list[item["corpus_id"]])
    
    #On fait un top 10 des réponses les plus probables et qui sont le plus probable de sortir (arrondi >0.6)
    if len(output_text) < 10 and score >= 0.6:
        output_text.append(unique_list[item["corpus_id"]])

print(output_text)
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import time
from datetime import datetime
from csv import writer

t0 = time.time()
model_name = "Qwen/Qwen1.5-1.8B-Chat"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)


prompt_template = """
Contexte : Voici une liste d'extraits de document de référence, sélectionnés comme étant les plus pertienents selon la question. Utilise uniquement ces passages pour répondre.
Document fournis : """ + str(output_text) +  """\n
Consignes : \n
- Ne génère ta réponse qu'en te basant strictement sur le contenu des documents ci-dessus.
- Si les documents ne permettent pas de répondre clairement à la question, dis simplement : 
"Je ne peux pas répondre à cette question avec les documents fournis."
Question :
"""+ user_text  +"""
"""
if len(output_text) > 0:
    #On remplace les mots par des vecteurs : 
    inputs = tokenizer(prompt_template,return_tensors="pt")
    eos_token_id = tokenizer.eos_token_id  # Utilise le token EOS pour arrêter la génération

    # Permet de faire répondre le modèle sur les vecteurs que l'on vient de créer
    outputs = model.generate(**inputs, temperature=0.1, max_new_tokens=200, eos_token_id=eos_token_id)

    #On décode la réponse renvoyée par le modèle : 
    response = tokenizer.decode(outputs[0],skip_special_tokens=True)
    print(response)
else:
    response = "Désolé, les fichiers ne permettent pas de répondre à vos questions."
t1 = time.time()
total_time = round(t1-t0,5)

#Inscription de la date : 
date_actuelle = datetime.now().strftime("%d/%m, %H:%M:%S")
# Inscription dans le fichier txt : 
with open("data_saved.txt", "a", newline="", encoding="utf-8") as fichier:
    fichier.write("-------------------------------------------------------------------\n")
    fichier.write("Test du : " + date_actuelle + " avec modèle : " + model_name + "\n")
    fichier.write("-------------------------------------------------------------------\n")
    fichier.write("Temps d'éxecution : " + str(total_time) + "\n")
    fichier.write("Question posée : " + user_text + "\n")
    fichier.write("Réponse apportée : " + response + "\n")
    fichier.write("===================================================================\n\n")
fichier.close()

#Pareil mais en csv : 
dictio_csv = [model_name,date_actuelle,total_time]
with open("data_saved.csv","a",encoding="utf-8",newline="") as fichier:
    w = writer(fichier)
    w.writerow(dictio_csv)
    fichier.close()
```
